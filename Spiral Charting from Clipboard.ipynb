```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_clipboard(sep='\t')
data
```

```python
#Drop Rows
data2=data.drop([0,1])
data2
```

```python
#Drop Columns
data3=data2.drop(['Unnamed: 1','Unnamed: 2'],axis=1)
data3
```

```python
data4=data3[~data3['Unnamed: 0'].str.contains('Naphtha|ZZ|Sulfur|Coke|Loss|Ammonia|Summary')]
data4
```

```python
data4=data4.set_index('Unnamed: 0')
data4
```

```python
data4=data4.astype(float)
data4=data4.sort_values('Base',ascending=False)
data4=data4.fillna(0)
data4=data4[(data4.T != 0).any()]
data4 = data4.loc[:,(data4 != 0).any(axis=0)]
data4.to_clipboard()
```

```python
import matplotlib.style as style
```

```python
#plot the data

#ax=plt.gca

#make a plot
#ax.plot(data4['Summary.Objective Function'])
style.use('seaborn-white')


data4.T.plot.bar(stacked=True, subplots=False, figsize=(15,6),sharey=True, width=1, edgecolor='black', rot=45, cmap='Set3')
#plt.legend(bbox_to_anchor=(1.05, 1.0),loc="lower center")

# Second axis data
data5=data3[~data3['Unnamed: 0'].str.contains('Naphtha|ZZ|Sulfur|Coke|Loss|Ammonia')]
data5=data5.set_index('Unnamed: 0')
data5=data5.astype(float)
data5=data5[(data5.T != 0).any()]
data5=data5.loc[:,(data5 != 0).any(axis=0)]
plt.legend(loc="lower center",bbox_to_anchor=(.5, -.3),ncol=7)
#plt.subplots_adjust(bottom=1.0)
#plt.show()



data5=data5.T
plt.ylabel("Rate (Mbpd)")
plt.xlabel("Mbpd")

data5['Summary.Objective Function'].plot(secondary_y=True, color='black', rot=45, linewidth=4,)
plt.ylabel("$MM/d")
plt.xlabel("Mbpd")
#plt.legend(loc="lower center",bbox_to_anchor=(1.0, 1.0))
#plt.subplots_adjust(hspace=1.05)


plt.legend(loc="lower center",bbox_to_anchor=(.5, -.3),ncol=7)
plt.show()
```
